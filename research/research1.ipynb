## Applying PCA Approach



#### Scaling the Orignal data 

```python
df1= pd.read_csv('../csv_data/stocks.csv')
```

```python
from sklearn.preprocessing import StandardScaler
features = df1.values
sc = StandardScaler()
X_scaled = sc.fit_transform(features)
```

```python
X_scaled[:5]
```

```python
X_scaled.shape
```

### Looking for best components for PCA

```python
from sklearn.decomposition import PCA

pca = PCA().fit(X_scaled)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
pca.components_
```

#### Apply PCA to reduce the number of dimensions from 502 to 2 dimensions for better data visualization.

```python
pca = PCA(n_components=2)
pca.fit(X_scaled)

X_transformed = pca.transform(X_scaled)

X_retransformed = pca.inverse_transform(X_transformed)
```

```python
X_transformed.shape
```

```python
X_retransformed.shape
```

## 2) Setting cluster on the Scaled Data using Kmeans++

```python
wcss=[]
for i in range(1, 21):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 121)
    kmeans.fit(X_transformed)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 21), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Mean Squared Errors')
plt.figure(figsize=(10,8))
plt.show()
```

#### AS we can see k=5 is optimum

#### Applying Kmeans clustering finally for k=5

```python
k_means_plus = KMeans(n_clusters=5,random_state=121,init='k-means++')
k_means_plus.fit(X_transformed)
y_kmeans = k_means_plus.fit_predict(X_transformed)
labels = k_means_plus.labels_
```
